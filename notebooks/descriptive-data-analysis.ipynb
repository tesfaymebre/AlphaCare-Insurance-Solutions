##### Data Structure:

```python
dq_checker.check_data_types(df)
```

```python
# Fix data type format in proper format
df['VehicleIntroDate']=pd.to_datetime(df['VehicleIntroDate'],errors='coerce')
df['TransactionMonth']=pd.to_datetime(df['TransactionMonth'], errors='coerce')
df['RegistrationYear']=(df['RegistrationYear']).astype(int)
```

```python
# filter numerical and catagorical columns for further processing
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object', 'bool']).columns.tolist()
```

### Data Quality Assessment

```python
# Display missing value report
dq_checker.check_missing_values()
```

```python
# Check is there is a duplicate values
dq_checker.check_duplicates()
```

```python
# cleaning the data based on the datatype
# if catagorical we use mode
for i in categorical_cols:
    if df[i].isnull().any():  # Check if there is any null value in the column
        df[i].fillna(df[i].mode()[0], inplace=True)  # Fill with mode
#if numerical we use mean
for i in numerical_cols:
    if df[i].isnull().any():  # Check if there is any null value in the column
        df[i].fillna(df[i].mean(), inplace=True)  # Fill with mean
```

```python
dq_checker.check_missing_values()
```

### Observations
- The overwhelming majority of vehicles are not VAT registered, indicating that most policies in this dataset are for vehicles that are not subject to VAT.
- ***Province***: The most policies are sold in Gauteng, followed by KwaZulu-Natal and the Western Cape. The least policies are sold in the Free State and Northern Cape.
- The overwhelming majority of vehicles are more than 6 months old, indicating that most policies in this dataset are for used vehicles, not new ones.
- The most popular vehicle type is "Passenger Vehicle," with a significantly higher count compared to the other vehicle types. This suggests that passenger vehicles are the most commonly insured type of vehicle in this dataset.
- The overwhelming majority of policies are monthly, indicating that monthly terms are significantly more popular than annual terms in this dataset.
- The most popular cover group is "Comprehensive - Taxi," with a significantly higher count compared to the other cover groups. This suggests that comprehensive coverage for taxis is the most sought-after type of insurance in this dataset.
- The most popular product is "Mobility Commercial Cover: Monthly," with a significantly higher count compared to the other products.
